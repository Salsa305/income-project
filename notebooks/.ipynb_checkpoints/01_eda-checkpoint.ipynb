## Income Label Distribution Analysis

The bar graph above illustrates the distribution of income categories within our dataset. The `<= 50k` category, representing individuals earning $50,000 or less, is significantly more prevalent than the `> 50K` category, indicating a substantial class imbalance.

### Implications of Class Imbalance

Class imbalance can affect the performance of machine learning models, potentially leading to biased predictions that favor the majority class. This can have serious consequences, especially in applications where the minority class is of particular interest.

### Planned Mitigation Strategies

To address the class imbalance and improve the robustness of our analysis, we will consider the following strategies:

1. **Metric Selection**: Utilize metrics such as precision, recall, F1 score, and ROC-AUC that provide a more nuanced view of model performance in imbalanced datasets.

2. **Data Sampling**: Explore techniques like oversampling the minority class, undersampling the majority class, or using SMOTE to generate synthetic samples.

3. **Algorithm Choice**: Investigate algorithms that are more resilient to class imbalance, such as tree-based methods and ensemble learning techniques.

4. **Cost-sensitive Learning**: Implement cost-sensitive learning to penalize misclassifications of the minority class more heavily.

5. **Feature Engineering**: Conduct a thorough feature analysis to identify potential new features or transformations that could improve model performance.

6. **Cross-validation**: Use cross-validation to ensure that our performance estimates are reliable and not influenced by the specific data split.

7. **Model Calibration**: Ensure that our model's probability estimates are well-calibrated, which is crucial for making informed decisions based on predictions.

By implementing these strategies, we aim to develop a model that is both accurate and fair, capable of making reliable predictions across both income categories.

```python
df = pd.DataFrame(X)
df['income'] = y.squeeze()

sns.set(style="whitegrid")

# Split numeric and categorical
numeric_features = df.select_dtypes(include=['int64', 'float64']).columns.drop('income')
categorical_features = df.select_dtypes(include=['object', 'category']).columns

# Detect skewed variables (|skew| > 1)
skewness = df[numeric_features].skew()
skewed_vars = skewness[abs(skewness) > 1].index.tolist()

print("Log-scaling the following skewed variables:")
print(skewed_vars)

# Apply log1p to skewed variables
df_log = df.copy()
for col in skewed_vars:
    df_log[col] = np.log1p(df_log[col])

# ---------------------------
# 1. Numeric Features
# ---------------------------
for feature in numeric_features:

    plt.figure(figsize=(10, 4))

    # Two overlapping histograms WITHOUT transparency
    sns.histplot(
        data=df_log[df_log['income'] == 0],
        x=feature,
        bins=30,
        element="bars",
        alpha=1.0,          # solid color
        linewidth=1,        # outline for clarity
        label="Income = 0",
        color="tab:blue"    # consistent coloring
    )

    sns.histplot(
        data=df_log[df_log['income'] == 1],
        x=feature,
        bins=30,
        element="bars",
        alpha=1.0,          # solid color
        linewidth=1,
        label="Income = 1",
        color="tab:orange"
    )

    # Titles
    if feature in skewed_vars:
        plt.title(f"Distribution of {feature} by Income (LOG-SCALED)")
        plt.xlabel(feature + " (log1p)")
    else:
        plt.title(f"Distribution of {feature} by Income")

    plt.ylabel("Count")
    plt.legend()
    plt.tight_layout()
    plt.show()

# ---------------------------
# 2. Categorical Features
# ---------------------------
for feature in categorical_features:

    plt.figure(figsize=(12, 4))
    sns.countplot(
        data=df,
        x=feature,
        hue="income"
    )

    plt.title(f"Distribution of {feature} by Income")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout()
    plt.show()
```

#  Updated Feature Distribution Summary (Matching the New Graphs)

### Age Distribution

* **Description:** The age distribution is smooth and unimodal, peaking in the early 30s.
  When separated by income class, individuals earning 50K dominate across all ages, while those earning >50K increase noticeably from the mid-30s to early 50s.
* **Implication:** Age remains an important predictor of income, with higher-income individuals tending to fall within a narrower mid-career age range.

---

### Final Weight (fnlwgt)

* **Description:** After applying **log scaling**, the extreme right skew is reduced and the distribution becomes much more interpretable.
  The two income groups overlap heavily, but subtle differences in density become visible.
* **Implication:** Log transformation helps stabilize variance and makes fnlwgt more usable in models sensitive to skewness (e.g., linear models).

---

### Education Level (education-num)

* **Description:** The distribution is discrete with clear peaks at common education levels (e.g., HS graduate, some college, bachelors).
  Income groups separate visually: higher education levels correspond to a larger fraction of >50K earners.
* **Implication:** Education-num is one of the strongest predictors of income, and the discrete structure is both interpretable and model-friendly.

---

### Capital Gain and Capital Loss

* **Description:** Both variables are **heavily zero-inflated**, and after **log transformation**, the non-zero values become visible with a long-tailed distribution.
  Individuals earning >50K show noticeably more non-zero capital gains.
* **Implication:** These features behave more like *binary-plus-continuous* variables.
  Creating flags like `has_capital_gain` or `has_capital_loss` may help models capture the strong signal in the non-zero cases.

---

### Hours Per Week

* **Description:** Hours-per-week shows a concentrated spike at **40 hours**, with secondary peaks at typical part-time and overtime values.
  After log scaling, the right tail becomes easier to see, but the main structural pattern remains.
* **Implication:** Work hours correlate with income, especially in the 4060 hour range, where >50K earners are more frequent.

```python
# Assuming df is your DataFrame and y is your processed target variable
df = pd.DataFrame(X)
df['income'] = y

# Randomly sample 2500 rows from the DataFrame
df_sampled = df.sample(n=2500, random_state=42)

# Bin the age and hours-per-week features for better visualization
df_sampled['age_bins'] = pd.cut(df_sampled['age'], bins=10)
df_sampled['hours_bins'] = pd.cut(df_sampled['hours-per-week'], bins=10)

# Select numeric columns for pairplot
numeric_cols = ['age', 'fnlwgt', 'education-num', 'capital-gain', 'capital-loss', 'hours-per-week']

# Create a pairplot for numeric variables
plt.figure(figsize=(12, 10))
sns.pairplot(df_sampled, hue='income', vars=numeric_cols)
plt.suptitle('Pairwise Relationships for Numeric Variables', y=1.02)
plt.show()

# Create bar plots for categorical variables
categorical_vars = ['age_bins', 'hours_bins']

for var in categorical_vars:
    plt.figure(figsize=(8, 6))
    sns.countplot(data=df_sampled, x=var, hue='income')
    plt.title(f'Income Distribution by {var}')
    plt.xticks(rotation=45)
    plt.show()

# Create histograms for each numeric variable
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.histplot(df_sampled, x=col, kde=True, hue='income', multiple='stack')
    plt.title(f'Distribution of {col}')
    plt.show()

# Create box plots for each numeric variable
for col in numeric_cols:
    plt.figure(figsize=(6, 4))
    sns.boxplot(x='income', y=col, data=df_sampled)
    plt.title(f'Box Plot of {col} by Income')
    plt.show()
```
